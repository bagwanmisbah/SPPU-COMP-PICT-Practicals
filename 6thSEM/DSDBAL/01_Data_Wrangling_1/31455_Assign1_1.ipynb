## Handling missing values 

### YearBuilt attribute has missing values/ Nan that can be filled with mean 

```python
mean_year=df['YearBuilt'].mean()
df['YearBuilt']=df['YearBuilt'].fillna(mean_year)
```

```python
mean_year
```

#### the Mean Year is 1964 and will repace null values

#### converting Float value of YearBuilt to int

```python
df['YearBuilt']=df['YearBuilt'].astype(int)
```

```python
df['YearBuilt'].head(3)
```

```python
df.isna().sum()  
# 62 null car values before 
# 6450 BuildingArea null values
# 1369 CouncilArea null values
```

```python
df['Car'] = df['Car'].ffill()
df['BuildingArea'] = df['BuildingArea'].bfill()
df['CouncilArea'] = df['CouncilArea'].ffill()
```
